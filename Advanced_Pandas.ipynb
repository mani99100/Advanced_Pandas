```python
# Q2. Sales Filtering Analysis
import pandas as pd

# Sample DataFrame
data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Region": ["North", "East", "West", "East", "South", "East"],
    "Sales": [4000, 2800, 3600, 1500, 5000, 3000]
}
df = pd.DataFrame(data)

# 1. Orders where Sales > 3500
print(df[df["Sales"] > 3500])

# 2. Orders where Region = "East" AND Sales > 2500
print(df[(df["Region"] == "East") & (df["Sales"] > 2500)])
```

```python
# Q3. Customer Purchase Analysis
import pandas as pd

# Sample DataFrame
data = {
    "Customer": ["A", "B", "C", "A", "D", "B", "E", "C", "F", "A"],
    "Sales": [2000, 3500, 1500, 4000, 2500, 3000, 4500, 5000, 1000, 6000]
}
df = pd.DataFrame(data)

# Group by Customer and calculate total sales
customer_sales = df.groupby("Customer")["Sales"].sum().reset_index()

# Sort by Sales in descending order and get top 5
top_customers = customer_sales.sort_values(by="Sales", ascending=False).head(5)
print(top_customers)
```

```python
# Q4. Regional Sales Performance
import pandas as pd

# Sample DataFrame
data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Region": ["North", "East", "West", "East", "South", "West"],
    "Sales": [4000, 2800, 3600, 1500, 5000, 3000]
}
df = pd.DataFrame(data)

# Calculate total sales by region
region_sales = df.groupby("Region")["Sales"].sum().reset_index()
print(region_sales)

# Identify region with highest revenue
highest_region = region_sales.loc[region_sales["Sales"].idxmax()]
print("Highest Revenue Region:", highest_region["Region"], "with Sales:", highest_region["Sales"])
```

```python
# Q5. Product Performance Analysis
import pandas as pd

# Sample DataFrame
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Laptop", "Camera", "Mobile", "Printer", "Tablet"],
    "Quantity": [5, 8, 4, 7, 6, 9, 3, 5]
}
df = pd.DataFrame(data)

# Calculate total quantity sold for each product
product_qty = df.groupby("Product")["Quantity"].sum().reset_index()
print(product_qty)

# Identify most popular product
popular_product = product_qty.loc[product_qty["Quantity"].idxmax()]
print("Most Popular Product:", popular_product["Product"], "with Quantity:", popular_product["Quantity"])
```

```python
# Q6. Multi-Level Grouping
import pandas as pd

# Sample DataFrame
data = {
    "OrderID": [101, 102, 103, 104, 105, 106, 107],
    "Region": ["North", "East", "West", "East", "South", "West", "North"],
    "Category": ["Electronics", "Furniture", "Electronics", "Furniture", "Clothing", "Clothing", "Electronics"],
    "Sales": [4000, 2800, 3600, 1500, 5000, 3000, 2500]
}
df = pd.DataFrame(data)

# Group by Region and Category, calculate total sales
grouped_sales = df.groupby(["Region", "Category"])["Sales"].sum().reset_index()
print(grouped_sales)
```

```python
# Q7. Pivot Table Creation
import pandas as pd

# Sample DataFrame
data = {
    "Region": ["North", "East", "West", "East", "South", "West", "North", "East"],
    "Product": ["Laptop", "Laptop", "Mobile", "Laptop", "Camera", "Laptop", "Laptop", "Mobile"],
    "Sales": [4000, 2800, 3600, 1500, 5000, 3000, 2500, 2000]
}
df = pd.DataFrame(data)

# Create pivot table
pivot = pd.pivot_table(df, values="Sales", index="Region", columns="Product", aggfunc="sum", fill_value=0)
print(pivot)

# Identify region with highest Laptop sales
highest_laptop_region = pivot["Laptop"].idxmax()
print("Region with highest Laptop sales:", highest_laptop_region)
```

```python
# Q8. Profitability Analysis
import pandas as pd

# Sample DataFrame
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Camera", "Printer", "Headphones"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Office", "Electronics"],
    "Profit": [2000, 3500, 1500, 2500, 800, 1200]
}
df = pd.DataFrame(data)

# Calculate average profit for each category
category_profit = df.groupby("Category")["Profit"].mean().reset_index()
print(category_profit)

# Identify most profitable category
most_profitable = category_profit.loc[category_profit["Profit"].idxmax()]
print("Most Profitable Category:", most_profitable["Category"], "with Avg Profit:", most_profitable["Profit"])
```

```python
# Q9. Discount Impact Study
import pandas as pd

# Sample DataFrame
data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Discount": [0.10, 0.20, 0.30, 0.10, 0.40, 0.20],
    "Profit":   [2000, 1500, 800, 2200, 500, 1200]
}
df = pd.DataFrame(data)

# Calculate average profit for each discount level
discount_profit = df.groupby("Discount")["Profit"].mean().reset_index()
print(discount_profit)

# Check relationship: do higher discounts reduce profit?
trend = discount_profit.sort_values("Discount")
print(trend)
```

```python
# Q10. End-to-End Data Wrangling Task
import pandas as pd

# Sample DataFrame
data = {
    "OrderID": [101, 102, 103, 104, 105, 106, 107],
    "Region": ["North", "East", "West", "East", "South", "West", "North"],
    "Product": ["Laptop", "Laptop", "Mobile", "Camera", "Laptop", "Laptop", "Camera"],
    "Sales": [4000, 2800, 3600, 1500, 5000, 3000, 2500]
}
df = pd.DataFrame(data)

# 1. Filter dataset where Sales > 3000
filtered = df[df["Sales"] > 3000]

# 2. Group by Region and Product, calculate total sales
grouped = filtered.groupby(["Region", "Product"])["Sales"].sum().reset_index()

# 3. Reshape into pivot table
pivot = pd.pivot_table(grouped, values="Sales", index="Region", columns="Product", aggfunc="sum", fill_value=0)
print(pivot)
```
